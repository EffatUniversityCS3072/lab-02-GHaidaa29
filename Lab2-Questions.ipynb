#### Practice Exercise 1

Create the following NumPy arrays: 
1. A 1-D array called zeros having 10 elements and all the elements are set to zero. 
2. A 1-D array called vowels having the elements ‘a’, ‘e’, ‘i’, ‘o’ and ‘u’. 
3. A 2-D array called ones having 2 rows and 5 columns and all the elements are set to 1 and dtype as int. 
4. Use nested Python lists to create a 2-D array called myarray1 having 3 rows and 3 columns and store the following data: 2.7,  0,  -2,  3.4,  10.6,  0,  -19 99.9 13 
5. A 2-D array called myarray2 using arange() having 3 rows and 5 columns with start value = 4, step size 4 and dtype as float


```python
import numpy as np

# 1. A 1-D array called zeros having 10 elements and all elements set to zero.
zeros = np.zeros(10)

# 2. A 1-D array called vowels having the elements ‘a’, ‘e’, ‘i’, ‘o’ and ‘u’.
vowels = np.array(['a', 'e', 'i', 'o', 'u'])

# 3. A 2-D array called ones having 2 rows and 5 columns, all elements set to 1, dtype as int.
ones = np.ones((2, 5), dtype=int)

# 4. A 2-D array called myarray1 using nested Python lists.
myarray1 = np.array([[2.7, 0, -2], [3.4, 10.6, 0], [-19, 99.9, 13]])

# 5. A 2-D array called myarray2 using arange() with specified parameters.
myarray2 = np.arange(4, 4 * 15, 4).reshape(3, 5).astype(float)

# Display the results
print("zeros:", zeros)
print("vowels:", vowels)
print("ones:\n", ones)
print("myarray1:\n", myarray1)
print("myarray2:\n", myarray2)


```

```python
# Find dimensions, shape, size, data type, and itemsize of the arrays
arrays = {
    'zeros': zeros,
    'vowels': vowels,
    'ones': ones,
    'myarray1': myarray1,
    'myarray2': myarray2
}

for name, array in arrays.items():
    print(f"{name}:")
    print(f"  Dimensions: {array.ndim}")
    print(f"  Shape: {array.shape}")
    print(f"  Size: {array.size}")
    print(f"  Data type: {array.dtype}")
    print(f"  Itemsize: {array.itemsize} bytes\n")

# Reshape the array ones to have all 10 elements in a single row
ones_reshaped = ones.reshape(1, -1)
print("Reshaped ones:\n", ones_reshaped)

# Display the 2nd and 3rd element of the array vowels
print("2nd and 3rd elements of vowels:", vowels[1:3])

# Display all elements in the 2nd and 3rd row of the array myarray1
print("2nd and 3rd rows of myarray1:\n", myarray1[1:3])

# Display the elements in the 1st and 2nd column of the array myarray1
print("1st and 2nd columns of myarray1:\n", myarray1[:, 0:2])

# Display the elements in the 1st column of the 2nd and 3rd row of the array myarray1
print("1st column of the 2nd and 3rd row of myarray1:", myarray1[1:3, 0])

# Reverse the array of vowels
vowels_reversed = vowels[::-1]
print("Reversed vowels:", vowels_reversed)

```

```python
import numpy as np

# Create a 2-D array called myarray4 with 14 rows and 3 columns
myarray4 = np.arange(-1, -1 + 14 * 0.25, 0.25).reshape(14, 3)
print("myarray4:\n", myarray4)

# Split this array row-wise into 3 equal parts
myarray4_split = np.array_split(myarray4, 3)
for i, part in enumerate(myarray4_split):
    print(f"\nPart {i + 1}:\n", part)


```

#### Practice Exercise 2
Read the coordinates of the capital cities of the world from the country-capital-lat-long-population.csv file. Use NumPy to print the name and coordinates of the capital city closest to the Saudi Arabia capital - Riyadh.

Note that:

1. The Country Name for Saudi Arabia is given as Saudi Arabia in the data.
2. The ‘closeness’ of capital cities from the Saudi Arabia capital is based on the Euclidean distance of their coordinates to those of the Saudi Arabia capital.


Hints:

1. Use read_html() from the Pandas library to read the table.
2. Use the to_numpy() function of the Pandas DataFrame class to convert a DataFrame to a Numpy array
3. Use broadcasting to compute the euclidean distance of capital cities from Riyadh.

```python
import pandas as pd
import numpy as np

# Load the data
df = pd.read_csv("country-capital-lat-long-population.csv")

# Get the coordinates of Riyadh
riyadh_coords = df[df['Country'] == 'Saudi Arabia'][['Latitude', 'Longitude']].to_numpy()[0]

# Extract the coordinates of all capital cities
coordinates = df[['Capital City', 'Latitude', 'Longitude']].to_numpy()

# Calculate the Euclidean distance
distances = np.sqrt((coordinates[:, 1].astype(float) - riyadh_coords[0])**2 + 
                    (coordinates[:, 2].astype(float) - riyadh_coords[1])**2)

# Find the index of the closest capital city
closest_index = np.argmin(distances)

# Get the closest capital city name and its coordinates
closest_capital = coordinates[closest_index]
closest_capital_name = closest_capital[0]
closest_capital_coords = closest_capital[1:]

print(f"The closest capital city to Riyadh is {closest_capital_name} at coordinates {closest_capital_coords}.")

```

#### Practical Exercise 2

Use NumPy to:

1. Print the names of the countries of the top 10 capital cities closest to the US capital - Washington DC.

2. Create and print a NumPy array containing the coordinates of the top 10 cities.

```python
import pandas as pd
import numpy as np

# Load the data
df = pd.read_csv("country-capital-lat-long-population.csv")

# Get the coordinates of Washington, D.C.
dc_coords = df[df['Capital City'] == 'Washington, D.C.'][['Latitude', 'Longitude']].to_numpy()[0]

# Extract the coordinates of all capital cities along with their countries
coordinates = df[['Country', 'Capital City', 'Latitude', 'Longitude']].to_numpy()

# Calculate the Euclidean distance to Washington, D.C.
distances = np.sqrt((coordinates[:, 2].astype(float) - dc_coords[0])**2 + 
                    (coordinates[:, 3].astype(float) - dc_coords[1])**2)

# Get the indices of the top 10 closest capital cities
top_10_indices = np.argsort(distances)[:10]

# Extract the top 10 closest capital cities and their coordinates
top_10_countries = coordinates[top_10_indices, 0]  # Country names
top_10_coordinates = coordinates[top_10_indices, 2:4]  # Coordinates (Latitude, Longitude)

# Print the names of the countries
print("Top 10 capital cities closest to Washington, D.C.:")
for country in top_10_countries:
    print(country)

# Print the NumPy array containing the coordinates of the top 10 cities
print("\nCoordinates of the top 10 capital cities:")
print(top_10_coordinates)

```
